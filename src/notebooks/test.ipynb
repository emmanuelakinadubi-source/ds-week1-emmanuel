```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
# categorical variables

for c in ['Sex', 'Pclass', 'Embarked']:
    plt.figure()
    sns.countplot(x=c, data=df, order=df [c].value_counts().index)
    plt.title(f'{c} Distribution')
    plt.show()
```

```python
# Bivariate Analysis
```

```python
target = 'Survived'
df[target].value_counts()
```

```python
sns.barplot(x=target, y='Age', data=df)
plt.title('Age vs Survived')
plt.show()
```

```python
sns.boxplot(x=target, y='Age', data=df)
plt.title('Age vs Survived')
plt.show()
```

```python
# Multivariate Analysis

sns.catplot(x=target, y='Age', hue='Pclass', kind='bar', data=df)
plt.title('Age vs Survived by Passenger Class')
plt.show()
```
